```
import copy
print("Mean\n")

# function to calculate the row wise mean
def rowWiseMean(scores):
  mean_list = []
  for row in scores:
    sum = 0
    for i in row:
      sum += i
    mean_list.append(sum/len(row))
  return mean_list

# printing the row wise mean
print("row wise mean:")
print(rowWiseMean(scores))


# function to calculate column wise mean
def columnWiseMean(scores):
  mean_list = [0] * len(scores[0]) # a list initialised with value zero and of the same size as number of columns in scores
  for row in scores:
    i=0
    while (i < len(scores[0])):
      mean_list[i] += row[i] #taking sum of each corresponding column in each row
      i +=1
  mean_list = [x/len(scores) for x in mean_list] # using list comprehension to divide the sum in each column by number of rows
  return mean_list

# printing the column wise mean
print("Column wise mean:")
print(columnWiseMean(scores))

# function to calculate overall mean
def overallMean(scores):
  mean = 0
  for row in scores:
    for val in row:
      mean += val
  
  mean = mean / (len(scores) * len(scores[0]))
  return mean

# printing the overall mean
print("overall mean:")
print(overallMean(scores))


print('\n Median:\n')

# function to calculate row wise median
def rowWiseMedian(vali):
  median_list = []
  temp = list(vali)
  for row in temp:
    row.sort()
    if (len(row)%2):
      median_list.append(row[(len(row) - 1)/2])
    else:
      first_middle = int((len(row) - 1)/2)
      median = (row[first_middle] + row[first_middle + 1])/2 #finding the mean of the middle two elements
      median_list.append(median)
  return median_list

# printing the row wise median
print("Row wise median:")
print(rowWiseMedian(copy.deepcopy(scores)))


#function to calculate column wise median
def columnWiseMedian(vali):
  # creating a transpose of the 2D list
  temp = np.transpose(vali)

  median_list = []
  for row in temp:
    row.sort()
    if (len(row)%2):
      median_list.append(row[(len(row)-1)/2])
    else:
      first_middle = int((len(row) - 1)/2)
      median = (row[first_middle] + row[first_middle + 1])/2
      median_list.append(median)
  return median_list
  
#printing the column wise median
# first element of the list represent the median of the first column and so on
print("Column wise median:")
print(columnWiseMedian(copy.deepcopy(scores)))


#function to calculate overall median
def overallMedian(vali):
  temp = [0] * (len(vali) * len(vali[0]))

  for row in range(len(vali)):
    for col in range(len(vali[0])):
      temp.append(vali[row][col])
  
  temp.sort()
  if (len(temp) % 2):
    index = int((len(temp) - 1)/2)
    return temp[index]
  else:
    first_middle = int((len(temp) - 1)/2)
    return ((temp[first_middle] + temp[first_middle + 1])/2)

#printing the overall median
print("Overall median:")
print(overallMedian(copy.deepcopy(scores)))


# Mode
from scipy.stats import mode
print("\n Mode:\n")

#function to calculate row wise mode
def rowWiseMode(vali):
  temp = []
  for row in vali:
    m = mode(row)[0][0] #returns a numpy array (which is 2D in nature)
    temp.append(m)
  
  return temp


#printing the rowise Mode
print("Row wise mode:")
print(rowWiseMode(scores))


#function to calculate column wise mode
def columnWiseMode(vali):
  temp = []
  temp = np.transpose(vali) #converting columns to rows for easy calculation
  mode_list = []
  for row in temp:
    m = mode(row)[0][0]
    mode_list.append(m)
  
  return mode_list

# printing the column wise mode
print('Column wise Mode:')
print(columnWiseMode(scores))

#function to calculate overall mode
def overallMode(vali):
  temp = []
  for row in vali:
    for val in row:
      temp.append(val)
  overall_mode = mode(temp)[0][0] #mode() returns a numpy array

  return overall_mode

#printing the overall mode
print("Overall mode:")
print(overallMode(scores))

# Standard Deviation

print('\nStandard Deviation:\n')

# function to calculate row wise standard deviation

def rowWisedeviation(vali):
  std_deviation_list = []
  for row in vali:
    std_deviation_list.append(np.std(row))
  
  return std_deviation_list

# printing the row wise standard deviation
print("Row wise standard deviation:")   
print(rowWisedeviation(scores))


#function to calculate column wise standard deviation
def columnWisedeviation(vali):
  temp = np.transpose(vali) #transposing a 2D list
  std_deviation_list = []
  for row in temp:
    std_deviation_list.append(np.std(row))

  return std_deviation_list

#printing the column wise standard deviation
print('Column wise standard deviation:')
print(columnWisedeviation(scores))


# function to calculate the overall standard deviation
def overallDeviation(vali):
  temp = []
  for row in vali:
    for val in row:
      temp.append(val)
  
  return (np.std(temp))

#printing the overall standard deviation
print("Overall Standard deviation:")
print(overallDeviation(scores))
```

```
import pandas as pd
import numpy as np
df = pd.read_csv('wine.csv', delimiter = ',',usecols= ['Alcohol', 'Proline']) #importing only Alcohol and Proline column from CSV file into dataframe
print(df)
```

```
temp = df.values.tolist()  #converting the dataframe to 2D list and extracting the values
data = np.transpose(temp)  #transposing the new list created from dataframe to store all the values for alcohol in a single row and all the values for Proline in a single row
```

```
import matplotlib.pyplot as plt
plt.scatter(data[0], data[1], color = 'green') #using matplotlib library to generate a scatter plot of Alcohol vs Proline
plt.title('Scatter Plot') # the scatter plot will help us visualse the extent of linear correleation between Alcohol and Proline before finding the Pearson's correlation cofficient
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```

```
#now we will find the pearson correleation using scipy library pearsonr() function
from scipy.stats import pearsonr

corr = pearsonr(data[0], data[1]) # returns (correlation cofficient, p-value)

print("Pearson correlation coficient is: ", end='')
print(corr[0]) #corr[0] gives the correlation cofficient
```

```
#loading five columns (features) from the wine.csv dataset

features = ['Alcohol', 'Phenols', 'Flavanoids', 'Hue', 'Proline']

df = pd.read_csv('wine.csv', delimiter = ',', usecols= features)
print(df)
```

```
# converting the data frame to a 2D list
temp = df.values.tolist()
data = np.transpose(temp) #first row represent the data of the first column of dataframe ('Alcohol'), second row represents data of second column of dataframe ('Phenols') and so on
```

```
# generating the covariance matrix of the above matrix of 5 features using numpy library
mat = np.array(data)
cov_matrix = np.cov(mat)
print("Covariance Matrix is:")
for row in cov_matrix:
  for val in row:
    print("%f" % val, end=' ')
  print("\n")
```
